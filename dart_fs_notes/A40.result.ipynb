```python
df_result = pd.merge(df_items_all, df_corp_info_all[['corp_code', 'jurir_no', 'bizr_no']], 
                     how='left', left_on='corp_code', right_on='corp_code')
```

```python
# 금액을 백만원 단위로 변환
def change_val(v):
    if unit == '백만원':
        amt = v
    elif unit = '천원':
        amt = round(v / 1000, 0)
    elif unit = '원':
        amt = round(v / 1000000, 0)
    else:
        amt = 0
    return amt
```

```python
# 금액을 백만원 단위로 변환한 항목을 추가
df_result['m_val_a'] = df_result.apply(lambda x : change_val(x['val_a']), axis=1)
df_result['m_val_b'] = df_result.apply(lambda x : change_val(x['val_b']), axis=1)
```
