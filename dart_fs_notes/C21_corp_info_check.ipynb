```python
import os
import pandas as pd

# 파일 저장 위치
path_dir = "D:/PythonProject/data-gatherer/dart_fs_notes/company/purged/"
file_list = os.listdir(path_dir)

df_all = pd.DataFrame()    
for file_name in file_list:
    if file_name.startswith('corp_info_'):            
        df = pd.read_csv(path_dir + file_name, dtype=object)
        df_all = pd.concat([df_all, df])
    
df_all.drop_duplicates(subset=['corp_code'], keep='first', inplace=True, ignore_index=True)
df_all['induty_code'] = df_all['induty_code'].str.zfill(5)            
df_all.to_csv(path_dir + 'corp_info_all_test.csv', index=False)
```
