### 사업보고서 "재무제표 주석" 문서 URL 전체 목록과 사업보고서 목록과 Cross-check

```python
import pandas as pd
# 데이터 Loading
path_dir1 = "D:/PythonProject/data-gatherer/dart_fs_notes/fs_notes/"
df1 = pd.read_csv(path_dir1 + 'fs_notes_all.csv', dtype=object)
path_dir2 = "D:/PythonProject/data-gatherer/dart_fs_notes/items/"
df2 = pd.read_csv(path_dir2 + 'items_all.csv', dtype=object)
```

```python
# df2에 있는 corp_code 목록에 해당하는 df1의 목록을 설정
corp_codes = df2.loc[(df2.corp_cls == 'Y') & (df2.rpt_name.str[-8:-4] >= '2017') & (df2.rpt_name.str[-8:-4] <= '2021'), ]['corp_code'].unique()
df1 = df1[df1['corp_code'].isin(corp_codes)]
```

```python
# dart_rpt_all(df1)에는 있으나 fs_notes_all(df2)에는 데이터가 빠져있는 목록
df_missing = pd.merge(df1, df2, how='outer', left_on='rpt_num', right_on='rpt_num' , indicator=True).query("_merge=='left_only'")
```
