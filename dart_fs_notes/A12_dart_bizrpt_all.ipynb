### DART 정기공시('A') 사업보고서(최종) 목록 전체 내용 저장

* 사업보고서 목록 저장 파일의 내용을 수합해서 전체 목록을 csv 파일로 저장

```python
import os
import pandas as pd

import sys
module_path = "D:/PythonProject/data-gatherer/dart_fs_notes"
sys.path.append(module_path)
import mydart

# 사업보고서 파일 목록 위치
path_dir_bizrpt = os.path.join(mydart.root_path, "dart_bizrpt")
file_list = os.listdir(path_dir_bizrpt)

# 모든 파일을 읽어서 보고서 목록을 하나로 합한다.
df_all = pd.DataFrame()    
for file_name in file_list:
    if not file_name.startswith('bizrpt_all') and not file_name.startswith('__'):
        file_path = os.path.join(path_dir_bizrpt, file_name)
        df = pd.read_csv(file_path, dtype=object)
        df_all = pd.concat([df_all, df])
    
# 파일로 저장(파일이 있으면 Update)
bizrpt_file_path = os.path.join(path_dir_bizrpt, "bizrpt_all.csv")
df_all.to_csv(bizrpt_file_path, index=False)
```
