### DART 외부감사관련('F') 감사보고서(최종) 목록

* 보고서 접수일 기준 3개월 기간(start ~ end) / 최대 3개월 조회 가능함
* 감사보고서 목록을 csv 파일로 저장

```python
import OpenDartReader
from time import sleep

api_key = 'f2e08d4ed3de0ba3d5cbf59c04c223e02b1751a2'
dart = OpenDartReader(api_key)

years = range(2013, 2024)   # 보고서 접수일 (2013)년부터 (2024-1)년까지
dates = [{'start' : '0101', 'end' : '0331'},
         {'start' : '0401', 'end' : '0630'},
         {'start' : '0701', 'end' : '0930'},
         {'start' : '1001', 'end' : '1231'}]

# 파일 저장 위치
path_dir = "D:/PythonProject/data-gatherer/dart_fs_notes/dart_audit_rpt/"

str_expr = "report_nm.str.contains('감사보고서')"

for y in years:
    for p in dates:
        sleep(0.1)
        start_date = str(y) + p['start']
        end_date = str(y) + p['end']
        # 외부감사관련(kind='F') 보고서 최종(final=True)
        df_rpt = dart.list(start=start_date, end=end_date, kind='F', final=True)
        # 감사보고서
        df_biz_rpt = df_rpt.query(str_expr)
        # csv 파일로 저장
        file_name = 'F_' + start_date + '_' + end_date + '.csv'
        df_biz_rpt.to_csv(path_dir + file_name, index=False)
```
