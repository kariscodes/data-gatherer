### DART 정기공시('A') 사업보고서(최종) 목록 전체 내용 저장

* 사업보고서 목록 저장 파일의 내용을 수합해서 전체 목록을 csv 파일로 저장

```python
import os
import pandas as pd

# 파일 저장 위치
path_dir = "D:/PythonProject/data-gatherer/dart_fs_notes/dart_rpt/"
file_list = os.listdir(path_dir)

df_all = pd.DataFrame()    
for file_name in file_list:
    if file_name != 'A_rpt_all.csv' and not file_name.startswith('__'):
        df = pd.read_csv(path_dir + file_name, dtype=object)
        df_all = pd.concat([df_all, df])
    
df_all.drop_duplicates(subset=['corp_code','rcept_no'], keep='first', inplace=True, ignore_index=True)    
df_all.to_csv(path_dir + 'A_rpt_all.csv', index=False)
```

* 종목코드에 구분자 붙이기
> 'A' 유가증권, 'B' 코스닥, 'C' 코넥스, 'D' 기타

```python
df['corp_cls'].value_counts()
```

```python
cls_map = {'Y' : 'A', 'K' : 'B', 'N' : 'C', 'E' : 'D'}
def new_stock_code(cls, code):
    prefix = cls_map[cls]
    # print(str(code))
    if not str(code) == "nan":
        new_code = prefix + str(code)
    else:
        new_code = ""
    return new_code        
```

```python
df['stock_code2'] = df.apply(lambda x : new_stock_code(x['corp_cls'], x['stock_code']), axis=1)
```

* 연도별 사업보고서에 회사의 In/Out 현황
> 두 데이터프레임 비교

```python
df_in_all = pd.DataFrame()
df_out_all = pd.DataFrame()

for year in range(2017, 2024):
    df1 = df[df['report_nm'].str[-8:-4] == str(year-1)][['corp_code', 'corp_name']]
    df2 = df[df['report_nm'].str[-8:-4] == str(year)][['corp_code', 'corp_name']]
    # 전년도에 없었으나 해당 연도에 보고서가 있는 기업 (comes in)
    df_in = pd.merge(df1, df2, how='outer', indicator=True).query("_merge=='right_only'").drop(columns=['_merge'])
    df_in['in'] = str(year)
    df_in_all = pd.concat([df_in_all, df_in])
    # 전년도에 있었으나 해당 연도에 보고서가 없는 기업 (comes out)
    df_out = pd.merge(df1, df2, how='outer', indicator=True).query("_merge=='left_only'").drop(columns=['_merge'])
    df_out['out'] = str(year)
    df_out_all = pd.concat([df_out_all, df_out])
    
df_diff_all = pd.merge(df_out_all, df_in_all, how='outer')
```
