### DART 정기공시('A') 사업보고서(최종) 목록 전체 내용 저장

* 사업보고서 목록 저장 파일의 내용을 수합해서 전체 목록을 csv 파일로 저장

```python
import os
import pandas as pd

# 파일 저장 위치
path_dir = "D:/PythonProject/data-gatherer/dart_fs_notes/dart_rpt/"
file_list = os.listdir(path_dir)

df_all = pd.DataFrame()    
for file_name in file_list:
    if file_name != 'A_rpt_all.csv' and not file_name.startswith('__'):
        df = pd.read_csv(path_dir + file_name, dtype=object)
        df_all = pd.concat([df_all, df])
    
df_all.drop_duplicates(subset=['corp_code','rcept_no'], keep='first', inplace=True, ignore_index=True)    
df_all.to_csv(path_dir + 'A_rpt_all.csv', index=False)
```

### 파일에 저장된 사업보고서 목록 검색

> rcept_no : 접수번호

> corp_cls : 법인구분 Y(유가), K(코스닥), N(코넥스), E(기타)

> rm : 비고 상세 - 조합된 문자로 각각은 아래와 같은 의미가 있습니다.
* 유 : 유가증권 종목
* 코 : 코스닥 종목
* 채 : 채권 종목 채권상장법인 공시사항임
* 넥 : 코넥스 종목
* 공 : 공정거래 공시
* 연 : 연결 공시
* 정 : 이후 정정 보고 있음 (최종보고서가 아님)
* 철 : 철회된 보고서

```python
import pandas as pd
# 데이터 Loading
path_dir = "D:/PythonProject/data-gatherer/dart_fs_notes/dart_rpt/"
df = pd.read_csv(path_dir + 'A_rpt_all.csv', dtype=object)
print(df.shape[0])
df.head(3)
```

```python
# 보고서 결산년도 검색
year = '2017'
str_expr = "report_nm.str.contains(@year)"
df.query(str_expr)      
```

```python
# 보고서 결산년월 검색
year_month = '2017.12'
str_expr = "report_nm.str.contains(@year_month)"
df.query(str_expr)       
```

```python
# 연도별 보고서 현황 통계
df['report_nm'].str[-8:-4].value_counts()
```

```python
# 특정년도 보고서 결산년월 기업현황 통계
year = '2017'
str_expr = "report_nm.str.contains(@year)"
df.query(str_expr)['report_nm'].str[-8:-1].value_counts()
```

```python
# 기업명 검색
corp_name = '삼성전자'
str_expr = "corp_name.str.contains(@corp_name)"
df.query(str_expr).sort_values(by='rcept_no', ascending=True)      
```

```python
# 기업코드 검색
corp_code = '00126380'
str_expr = "corp_code.str.contains(@corp_code)"
df.query(str_expr).sort_values(by='rcept_no', ascending=True)     
```

```python
# 종목코드 검색
stock_code = '005930'
str_expr = "stock_code == @stock_code"
df.query(str_expr).sort_values(by='rcept_no', ascending=True)  
```

```python
# 상장종목 검색 (Y-유가증권, K-코스닥)
corp_cls = 'Y'
str_expr = "corp_cls.str.contains(@corp_cls)"
df.query(str_expr).sort_values(by='corp_code', ascending=True)   
```

```python
# 상장종목 + 결산년월 검색 (Y-유가증권, K-코스닥)
corp_cls = 'Y'
year_month = '2017.12'
str_expr = "corp_cls.str.contains(@corp_cls) and report_nm.str.contains(@year_month)"
df.query(str_expr).sort_values(by='corp_code', ascending=True) 
```
