```python
import requests
import xmltodict
import pandas as pd

service_url = 'http://apis.data.go.kr/1130000/appnGroupAffiList'
service_name = 'appnGroupAffiListApi'   # API
# 일반 인증키(Encoding)	
api_key = 'IVgu%2FZBjA6hpLryyEOpySC2RhogOhaJIUqlXN8Uyj3Gxw4s3dX0qMxfgXMTLl60%2Fs2EYAMUsyyzTqwVOnjoIhg%3D%3D'
ym = '202105'    # year month
rows = 10000
#url = f'{service_url}/{service_name}?ServiceKey={api_key}&presentnYear={ym}&unityGrupCode=K1000032&numOfRows=10&pageNo=1'
url = f'{service_url}/{service_name}?ServiceKey={api_key}&presentnYear={ym}&numOfRows={rows}&pageNo=1'
print(url)
```

```python
response = requests.get(url,verify=False)
if response.status_code == 200:     # URL GET '200 정상'
    print("URL GET Successful")
    contents = response.text                # XML 데이터
    # xmltodict 모듈을 사용해서 xml 데이터를 파이썬의 OrderedDict 타입으로 변환
    dict_data = xmltodict.parse(contents)
else:
    print("URL GET Error")
```
