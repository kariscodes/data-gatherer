```python
from bs4 import BeautifulSoup
import requests

url = "https://wkdus0608.tistory.com/entry/%EC%B2%B4%EC%9C%A1%ED%95%99%EA%B3%BC-%ED%95%99%EA%B3%BC-%EC%A0%95%EB%B3%B4-%EB%B0%8F-%EC%A7%84%EB%A1%9C-%EB%8C%80%ED%95%99-%EC%88%9C%EC%9C%84-%EA%B0%9C%EC%84%A4-%EB%8C%80%ED%95%99"
req = requests.get(url)
soup = BeautifulSoup(req.content.decode('utf8', 'replace'), "html.parser")
```

```python
import pandas as pd

rows_list = []

table = soup.find('table').find_next('table')
rows = table.find_all('tr')

for row in rows:    
    data = row.find_all('td')
    
    cnt = 0
    for d in data:
        cnt += 1
        if cnt == 1:
            region = d.get_text()
        elif cnt == 2:
            university = d.get_text()
            if d.find('a') == None:
                homepage = ""
            else:
                homepage = d.find('a')['href']
        elif cnt == 3:
            subject = d.get_text()  
            
    dict = {'지역': region, '대학교': university, '학과': subject, '홈페이지': homepage}
    rows_list.append(dict)
    
df = pd.DataFrame(rows_list)
```
