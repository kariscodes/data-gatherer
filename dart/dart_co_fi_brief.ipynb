```python
import requests

url = 'https://opendart.fss.or.kr/api/fnlttSinglAcnt.json'
api_key = 'f2e08d4ed3de0ba3d5cbf59c04c223e02b1751a2'
corp_code = '01208885'

# report code
# 1분기보고서 : 11013
# 반기보고서 : 11012
# 3분기보고서 : 11014
# 사업보고서 : 11011
params = {
    'corp_code': corp_code,
    'bsns_year': 2022,
    'reprt_code': 11011,
    'crtfc_key': api_key,
}

response = requests.get(url, params=params)
if response.status_code != 200:     # URL GET '200 정상'
    print("URL GET Error")
```

```python
# 데이터프레임 형태로 변환
import pandas as pd
df_fi = pd.json_normalize(response.json()['list'])
```

```python
df_fi.dtypes
```

```python
df_fi = df_fi.astype({'ord' : 'int'})
```

```python
df_fi
```

```python
# fs_div	개별/연결구분	CFS:연결재무제표, OFS:재무제표
# sj_div	재무제표구분	BS:재무상태표, IS:손익계산서
fs_div = "OFS"
sj_div = "BS"
df_fi_bs = df_fi.query('fs_div == @fs_div and sj_div == @sj_div')[['ord', 'account_nm', 
                                                        'thstrm_nm', 'thstrm_dt', 'thstrm_amount',
                                                        'frmtrm_nm', 'frmtrm_dt', 'frmtrm_amount',
                                                        'bfefrmtrm_nm', 'bfefrmtrm_dt', 'bfefrmtrm_amount']].sort_values('ord')
df_fi_bs
```

```python
# fs_div	개별/연결구분	CFS:연결재무제표, OFS:재무제표
# sj_div	재무제표구분	BS:재무상태표, IS:손익계산서
fs_div = "OFS"
sj_div = "IS"
df_fi_is = df_fi.query('fs_div == @fs_div and sj_div == @sj_div')[['ord', 'account_nm', 
                                                        'thstrm_nm', 'thstrm_dt', 'thstrm_amount',
                                                        'frmtrm_nm', 'frmtrm_dt', 'frmtrm_amount',
                                                        'bfefrmtrm_nm', 'bfefrmtrm_dt', 'bfefrmtrm_amount']].sort_values('ord')
df_fi_is = df_fi_is[['ord', 'account_nm', 'thstrm_nm', 'thstrm_amount', 'frmtrm_nm', 'frmtrm_amount', 'bfefrmtrm_nm', 'bfefrmtrm_amount']]
df_fi_is
```

```python
# 숫자의 소수점(,) 없애기
df_fi_is['thstrm_amount'] = df_fi_is['thstrm_amount'].apply(lambda x: x.replace(',', ''))
df_fi_is['frmtrm_amount'] = df_fi_is['frmtrm_amount'].apply(lambda x: x.replace(',', ''))
df_fi_is['bfefrmtrm_amount'] = df_fi_is['bfefrmtrm_amount'].apply(lambda x: x.replace(',', ''))
df_fi_is
```

```python
# 타입 변환
df_fi_is = df_fi_is.astype({'thstrm_amount' : 'float',
                            'frmtrm_amount' : 'float',
                            'bfefrmtrm_amount' : 'float'})
df_fi_is
```
