+ 작성: 임경호
+ 금융감독원 OPEN DART https://opendart.fss.or.kr/

```python
import requests

url = 'https://opendart.fss.or.kr/api/fnlttSinglAcntAll.json'
api_key = 'f2e08d4ed3de0ba3d5cbf59c04c223e02b1751a2'
corp_code = '01208885'
# CFS:연결재무제표, OFS:재무제표
fs_div = 'OFS'

# report code
# 1분기보고서 : 11013
# 반기보고서 : 11012
# 3분기보고서 : 11014
# 사업보고서 : 11011
params = {
    'corp_code': corp_code,
    'bsns_year': 2022,
    'reprt_code': 11011,
    'fs_div': fs_div,
    'crtfc_key': api_key,
}

response = requests.get(url, params=params)
if response.status_code != 200:     # URL GET '200 정상'
    print("URL GET Error")
```

```python
response.json()
```

```python
# 데이터프레임 형태로 변환
import pandas as pd
df_fi = pd.json_normalize(response.json()['list'])
```

```python
df_fi.dtypes
```

```python
df_fi = df_fi.astype({'ord' : 'int'})
```
