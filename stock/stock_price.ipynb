### FinanceDataReader
- pip install -U finance-datareader
- 사용자 안내서: https://financedata.github.io/posts/finance-data-reader-users-guide.html#
#### 종목 코드
- 거래소별 전체 종목코드: KRX (KOSPI, KODAQ, KONEX), NASDAQ, NYSE, AMEX, S&P 500
#### 가격 데이터
- 해외주식 가격 데이터: AAPL(애플), AMZN(아마존), GOOG(구글) 등
- 국내주식 가격 데이터: 005930(삼성전자), 091990(셀트리온헬스케어) 등
- 각종 지수: KS11(코스피지수), KQ11(코스닥지수), DJI(다우지수), IXIC(나스닥 지수), US500(S&P 5000)
- 환율 데이터: USD/KRX (원달러 환율), USD/EUR(달러당 유로화 환율), CNY/KRW: 위엔화 원화 환율
- 암호화폐 가격: BTC/USD (비트코인 달러 가격, Bitfinex), BTC/KRW (비트코인 원화 가격, 빗썸)

```python
# 네이버 금융
import FinanceDataReader as fdr

df_fdr = fdr.DataReader(stock_code, start=start_date, end=end_date)
df_fdr
```

### pykrx
- 국내주식 데이터만 조회 가능

```python
from pykrx import stock

# 코스피 종목 리스트
tickers = stock.get_market_ticker_list("20220222", market="KOSPI") 
```

```python
stock_name = stock.get_market_ticker_name("005930")
stock_name
```

```python
# 특정일 코스피 종목들의 OHLCV
ohlcv = stock.get_market_ohlcv("20220222", market="KOSPI") 
ohlcv
```

```python
# 특정기간 특정 종목의 OHLCV
ohlcv2 = stock.get_market_ohlcv("20220203", "20220222", "005930") 
ohlcv2
```
