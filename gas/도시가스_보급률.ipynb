### 파일 읽기: 도시가스 보급률
+ 작성: 임경호

```python
import pandas as pd

# 기준년도
base_year = '2021'
sheet = "('12-'21)"

# 파일 위치 (파일 출처: 도식가스협회 홈페이지 - 설비현황 자료 다운로드)
path = 'D:/PythonProject/data-gatherer/gas/data/5.+보급률실적+(1992~2021)_업로드.xlsx'

df_data = pd.read_excel(path, sheet_name=sheet, header=3)
```

```python
df_data
```

```python
# 첫번째와 두번째 컬럼만 추출
df_comp = df_data.iloc[:,0:2]
df_comp.columns = ['지역', '회사']
# 기준년도와 일치하는 컬럼만 추출
df_figure = df_data.filter(like=base_year, axis=1)
df_figure.columns = ['세대수', '수요가수', '보급률']
```

```python
# 두 개 데이터프레임 합치기
df_data = pd.concat([df_comp, df_figure], axis=1)
```

```python
df_data
```

```python
target_company = ['서   울 (계)', '삼천리 (계)', '대   성 (계)', '*대성(계)', 'CNCITY (계)', '경   동 (계)', '부   산']
df_target_data = df_data.query("지역 in @target_company or 회사 in @target_company")
df_target_data.shape
```

```python
# [지역] 컬럼의 값을 [회사] 컬럼으로 복사/붙여넣기
df_target_data.loc[:, ('회사')] = df_target_data['회사'].fillna(df_target_data['지역'])
# [지역] 컬럼 삭제
df_target_data = df_target_data.drop(['지역'], axis=1)
```

```python
df_target_data
```

```python
# 회사명에서 "(계)" 삭제: '(', ')'는 특수문자임
df_target_data.loc[:, ('회사')] = df_target_data['회사'].str.replace(pat='[^\w]계[^\w]', repl='', regex=True)
# 회사면의 공백 없애기
df_target_data.loc[:, ('회사')] = df_target_data['회사'].str.replace(pat=' ', repl='', regex=True)
```

```python
# 인덱스 초기화
df_target_data.reset_index(drop=True, inplace=True)
print(f"도시가스사 보급률(기준연도: {base_year})")
df_target_data
```
